## 使い方
### 実行結果内のURLにアクセス
### 日本語化
1. 画面上部「Settings」→左のメニュー欄「User Interface」→下部にある「Localization」のNoneを「ja JP」に変更
2. 上部にある「Apply settings」をクリック
3. その横にある「Reload UI」をクリック

### VAE設定
1. 画面上部「Settings」→左のメニュー欄「Stable Diffusion」→「SD VAE」のプルダウンメニューから使用するVAEを選ぶ
2. 上部にある「Apply settings」をクリック
